**2. Предобработка**

```python
PEAK_T = 150  # Целевая позиция пика

signal_line_corrected = np.zeros_like(signal_line)

# Выравниваем сигналы
for i, w in enumerate(signal_line):
    shift = PEAK_T - w.argmax()
    if shift == 0:
        signal_line_corrected[i] = w
    elif shift > 0:
        # Сдвиг вправо: обрезаем хвост, свободное место заполняется нулями (по умолчанию)
        signal_line_corrected[i, shift:] = w[:-shift]
    else:
        # Сдвиг влево: обрезаем начало, свободный конец остаётся нулевым
        signal_line_corrected[i, :shift] = w[-shift:]

# Лаконичная и надёжная проверка результата
peak_sample = signal_line_corrected.argmax(axis=1)
print(f"Целевая позиция: {PEAK_T}")
print(f"Все пики выровнены: {np.all(peak_sample == PEAK_T)}")
print(f"Уникальные позиции пиков: {np.unique(peak_sample)}")
```

```python
# Нормировка на амплитуду (приводим форму импульса к диапазону [0, 1])
peak_amps = signal_line_corrected.max(axis=1, keepdims=True)

# Защита от деления на ноль/шум: используем малый эпсилон вместо жёсткого порога
safe_amps = np.maximum(peak_amps, 1e-9)
normalized_waveforms = signal_line_corrected / safe_amps

# Векторизованная отрисовка
plt.figure(figsize=(12, 4))
plt.plot(normalized_waveforms[sel[:30]].T, alpha=0.2, linewidth=0.8, color='C0')
plt.title('Нормированные сигналы (сравнение формы импульсов)', fontsize=14)
plt.xlabel('Отсчёт, нс', fontsize=12)
plt.ylabel('Нормированная амплитуда', fontsize=12)
plt.xlim(120, 350)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()
```

**2. Вывод. Предобработка**

Произведена нормировка каждого импульса. Нормировка на максимум импульса переводит все зарегистрированные сигналы в единый масштаб [0,1]. Это позволяет отделить информацию о форме импульса (характеризующую тип частицы: гамма-квант или нейтрон) от информации об энергии (амплитуде). После нормировки алгоритмы кластеризации будут реагировать исключительно на различия во времени нарастания, спада и наличии "хвоста", что является основой метода импульсной дискриминации (PSD).

**3. Feature Engineering. Подбор признаков, их анализ и оценка важности**

**Проблема:** Исходные сигналы содержат 500 точек, что создаёт избыточную размерность для кластеризации. Требуется сжатие признакового пространства с сохранением информативности.

**Идея:** Если формы импульсов различаются между классами, то и распределение заряда по временны́м окнам будет различным. Это позволяет заменить сырой сигнал набором интегральных характеристик.

**Предлагаемые признаки** (интегралы относительно пика на позиции 150):
fast = 150..165 - быстрый компонент (начало импульса)
long = 165..249 - медленный компонент
tail = 205..249 - хвостовая часть
total = 145..259 - полный заряд

**Ключевые производные признаки:**
long / total - доля медленного компонента (основной дискриминант)
tail / total - относительная мощность хвоста
(long − fast) / total - контраст между ранней и поздней фазами
Ширина хвоста (например, по уровню 10% от максимума)
Наклон хвоста (линейная аппроксимация на участке 200..350)

**Преимущества подхода:**
Сокращение размерности: с 500 точек до 5-7 признаков
Физическая интерпретируемость признаков
Устойчивость к шуму за счёт интегрирования
Возможность масштабирования и нормировки для стабильной кластеризации

```python
def compute_feats(wn, wa, peak_idx=150, eps=1e-9):
    """
    Вычисляет физические признаки импульса с защитой от выбросов и деления на ноль.
    wn : np.ndarray
        Нормированный сигнал (для интегральных отношений)
    wa : np.ndarray
        Сигнал с коррекцией базовой линии (для временных метрик)
    peak_idx : int
        Индекс пика (по умолчанию 150)
    eps : float
        Эпсилон для безопасного деления
    """
    # Интегралы по окнам относительно пика
    qt    = wn[peak_idx-5 : peak_idx+99].sum()   # полный заряд
    qf    = wn[peak_idx   : peak_idx+15].sum()    # быстрая компонента
    ql    = wn[peak_idx+15: peak_idx+99].sum()   # медленная компонента
    qtail = wn[peak_idx+55: peak_idx+99].sum()   # дальний хвост

    # Защита от деления на ноль/отрицательный заряд
    qt_safe = max(abs(qt), eps)
    
    #r_long = ql / qt_safe
    r_long = ql / (qt + eps)
    r_tail = qtail / qt_safe
    r_diff = (ql - qf) / qt_safe

    # FWHM (ширина на полувысоте)
    pk = wa[peak_idx]
    half_thresh = 0.5 * pk
    above_half = np.where(wa >= half_thresh)[0]
    fwhm = float(above_half[-1] - above_half[0]) if len(above_half) > 1 else 0.0

    # Время спада (90% → 10% после пика)
    down = wa[peak_idx:]
    idx_90 = np.where(down >= 0.90 * pk)[0]
    idx_10 = np.where(down >= 0.10 * pk)[0]
    fall_t = float(idx_10[-1] - idx_90[-1]) if len(idx_10) > 0 and len(idx_90) > 0 else 0.0

    # Наклон логарифмического хвоста (экспоненциальный спад)
    seg = wn[peak_idx+20 : peak_idx+120]
    seg = np.clip(seg, eps, None)  # избегаем log(0) и отрицательных значений
    slope = float(np.polyfit(np.arange(seg.size), np.log(seg), 1)[0])

    return [r_long, r_tail, r_diff, fwhm, fall_t, slope]

# Пакетное применение
cols = ['r_long', 'r_tail', 'r_diff', 'fwhm', 'fall_t', 'slope']
feat_list = [compute_feats(wn, wa) for wn, wa in zip(normalized_waveforms, signal_line_corrected)]
feat = pd.DataFrame(feat_list, columns=cols)

# Корректная очистка артефактов
feat = feat.replace([np.inf, -np.inf], np.nan)
# Заполняем NaN медианой, но дополнительно клипируем физические выбросы (например, FWHM > 200 отсчётов)
feat = feat.fillna(feat.median())
feat[['fwhm', 'fall_t']] = feat[['fwhm', 'fall_t']].clip(upper=200.0)

# Статистика
display(feat.describe().round(4))
```

```python
log_amp = np.log10(peak_amp + 1)
corr, pval = pearsonr(log_amp, feat['r_long'])

fig, axes = plt.subplots(2, 1, figsize=(10, 8), height_ratios=[1, 1.5])

# Гистограмма с аннотацией
axes[0].hist(feat['r_long'], bins=200, color='steelblue', edgecolor='white', alpha=0.8)
axes[0].axvline(0, color='crimson', lw=1.5, ls='--', label='нулевая граница')
axes[0].set_title(f'Распределение r_long  |  μ={feat["r_long"].mean():.3f}, σ={feat["r_long"].std():.3f}')
axes[0].set_xlabel('r_long')
axes[0].legend(fontsize=9)
axes[0].grid(axis='y', alpha=0.3)

# Scatter + линия тренда + корреляция
axes[1].scatter(log_amp, feat['r_long'], s=3, alpha=0.1, color='darkslategray', label='события')
z = np.polyfit(log_amp, feat['r_long'], 1)
p = np.poly1d(z)
axes[1].plot(log_amp, p(log_amp), 'r--', lw=2, label=f'тренд: y={z[0]:.4f}x+{z[1]:.3f}')
axes[1].set_title(f'r_long vs амплитуда  |  Pearson r={corr:.3f} (p={pval:.2e})')
axes[1].set_xlabel('log10(амплитуда + 1)')
axes[1].set_ylabel('r_long')
axes[1].set_ylim(-0.3, 0.4)
axes[1].legend(fontsize=9)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Автоматически выбираем только числовые колонки
numeric_cols = feat.select_dtypes(include='number').columns.tolist()
print(f"Найдено числовых признаков: {len(numeric_cols)}")

# Функция расчёта корреляции с p-value
def corr_with_pvalue(df, col1, col2):
    mask = df[[col1, col2]].notna().all(axis=1)
    if mask.sum() < 3:  # Защита от ошибок при малом кол-ве точек
        return np.nan, np.nan
    
    res = pearsonr(df.loc[mask, col1], df.loc[mask, col2])
    # Совместимо со всеми версиями scipy (старые возвращают кортеж, новые - объект)
    corr = res.statistic if hasattr(res, 'statistic') else res[0]
    pval = res.pvalue if hasattr(res, 'pvalue') else res[1]
    return corr, pval

# Вычисляем попарные корреляции
significant_corrs = []
total_pairs = len(numeric_cols) * (len(numeric_cols) - 1) // 2
print(f" Вычисляю {total_pairs} парных корреляций...")

for i, col1 in enumerate(numeric_cols):
    for col2 in numeric_cols[i+1:]:
        corr, pval = corr_with_pvalue(feat, col1, col2)
        if pd.notna(pval) and pval < 0.05:
            significant_corrs.append({
                'feature_1': col1,
                'feature_2': col2,
                'pearson_r': corr,
                'p_value': pval,
                'abs_r': abs(corr)
            })

# Формируем итоговую таблицу
corr_df = pd.DataFrame(significant_corrs).sort_values('abs_r', ascending=False)
print("\nТоп-15 значимых корреляций (p < 0.05):")
print(corr_df[['feature_1', 'feature_2', 'pearson_r', 'p_value']].head(15).to_string(index=False))
```

**3. Вывод Feature Engineering. Подбор признаков, их анализ и оценка важности**

1. Ключевые извлеченные признаки:
**peak_amp** - амплитуда импульса (энергия частицы). Сильно скошено, логнормальное. Требует log10(amp+1) для нормализации.
**peak_pos** - позиция пика (время прихода). Узкое распределение вокруг медианы ~150 отсчётов.
**r_long** - параметр асимметрии формы импульса. Бимодальное распределение с границей около 0 — указывает на два класса событий.
**fall_t** - Время нарастания/спада. Коррелируют с амплитудой, важны для идентификации типа частицы.
**fwhm** - площадь импульса, ширина на полувысоте. Дополнительные метрики энергии и временны́х характеристик.
**slope** - (наклон) - это признак, который характеризует крутизну фронта импульса.
**r_diff** параметр дискриминации формы импулься
3. Корреляционная матрица признаков, значения в ней близкие к +-1 (например, между r_long и r_trail), указывают на сильную мультиколлинеарность, что делает такие признаки избыточными для кластеризации, так как они не добавляют новой информации. Напротив, признаки со слабой корреляцией (например, slope и fall_t) несут независимые характеристики формы сигнала. Но при этом сохранены все признаки в данной работе. 
4. Логарифмическое преобразование амплитуды улучшает линейность зависимостей и стабильность работы метрик кластеризации.



**4. Анализ выбросов**

```python
# 1. Подготовка признаков
X = feat.copy()

# Оставляем только числовые признаки
X = X.select_dtypes(include=[np.number])

# На всякий случай очищаем inf и NaN
X = X.replace([np.inf, -np.inf], np.nan)
X = X.fillna(X.median())

print("Размерность признаков:", X.shape)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Размерность после масштабирования:", X_scaled.shape)
```

```python
# 2. Isolation Forest
iso = IsolationForest(
    n_estimators=300,
    contamination=0.03,
    random_state=42
)

iso_labels = iso.fit_predict(X_scaled)

# -1 = выброс, 1 = нормальный сигнал
feat["is_outlier_iso"] = (iso_labels == -1).astype(int)

print("Количество выбросов Isolation Forest:")
print(feat["is_outlier_iso"].value_counts())
print("Доля выбросов:", feat["is_outlier_iso"].mean().round(4))
```

```python
# 3. Local Outlier Factor
lof = LocalOutlierFactor(
    n_neighbors=35,
    contamination=0.03
)

lof_labels = lof.fit_predict(X_scaled)

# -1 = выброс, 1 = нормальный сигнал
feat["is_outlier_lof"] = (lof_labels == -1).astype(int)

print("Количество выбросов LOF:")
print(feat["is_outlier_lof"].value_counts())
print("Доля выбросов:", feat["is_outlier_lof"].mean().round(4))
```

```python
# 4. Итоговая метка выброса
feat["outlier_votes"] = (
    feat["is_outlier_iso"] +
    feat["is_outlier_lof"]
)

# Мягкий вариант: выброс, если хотя бы один метод отметил сигнал
feat["is_outlier"] = (feat["outlier_votes"] >= 1).astype(int)

print("Итоговое количество выбросов:")
print(feat["is_outlier"].value_counts())
print("Итоговая доля выбросов:", feat["is_outlier"].mean().round(4))
```

```python
# 5. Визуализация выбросов в PCA-пространстве

pca = PCA(n_components=2, random_state=42)
X_pca_outliers = pca.fit_transform(X_scaled)

plt.figure(figsize=(9, 6))

sns.scatterplot(
    x=X_pca_outliers[:, 0],
    y=X_pca_outliers[:, 1],
    hue=feat["is_outlier"],
    palette="Set1",
    alpha=0.6,
    s=25
)

plt.title("Анализ выбросов в PCA-пространстве")
plt.xlabel("PCA component 1")
plt.ylabel("PCA component 2")
plt.legend(title="Выброс")
plt.grid(alpha=0.3)
plt.show()
```

```python
# 6. Сравнение признаков нормальных и аномальных сигналов
outlier_summary = feat.groupby("is_outlier")[cols].mean().T

outlier_summary.columns = ["normal_signals", "outliers"]

outlier_summary["difference"] = (
    outlier_summary["outliers"] - outlier_summary["normal_signals"]
)

outlier_summary["abs_difference"] = outlier_summary["difference"].abs()

outlier_summary.sort_values(
    by="abs_difference",
    ascending=False
)
```

```python
# 7. Распределения признаков для нормальных и аномальных сигналов
for col in cols:
    plt.figure(figsize=(8, 4))

    sns.kdeplot(
        data=feat,
        x=col,
        hue="is_outlier",
        fill=True,
        common_norm=False,
        alpha=0.4
    )

    plt.title(f"Распределение признака {col} для нормальных и аномальных сигналов")
    plt.xlabel(col)
    plt.ylabel("Density")
    plt.grid(alpha=0.3)
    plt.show()
```

```python
# 8. Визуализация нескольких аномальных сигналов
outlier_indices = feat[feat["is_outlier"] == 1].index[:20]

plt.figure(figsize=(12, 6))

for idx in outlier_indices:
    plt.plot(signal_line_corrected[idx], alpha=0.4)

plt.title("Примеры аномальных сигналов")
plt.xlabel("Time step")
plt.ylabel("Signal value")
plt.grid(alpha=0.3)
plt.show()
```

```python
# 9. Сравнение средней формы нормальных и аномальных сигналов
normal_idx = feat[feat["is_outlier"] == 0].index
outlier_idx = feat[feat["is_outlier"] == 1].index

normal_mean_signal = signal_line_corrected[normal_idx].mean(axis=0)
outlier_mean_signal = signal_line_corrected[outlier_idx].mean(axis=0)

plt.figure(figsize=(10, 5))

plt.plot(normal_mean_signal, label="Средний нормальный сигнал")
plt.plot(outlier_mean_signal, label="Средний аномальный сигнал")

plt.title("Сравнение средней формы нормальных и аномальных сигналов")
plt.xlabel("Time step")
plt.ylabel("Signal value")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

Для поиска аномальных сигналов были использованы два независимых алгоритма:
- Isolation Forest;
- Local Outlier Factor (LOF).

Каждый из методов выделил около 3% сигналов как аномальные. После объединения результатов итоговая доля выбросов составила 5.78%, что соответствует 1358 сигналам из 23479. Анализ средних значений признаков показал, что основное отличие аномальных сигналов связано с параметром `fall_t` — временем спада сигнала.

Среднее значение:
- для нормальных сигналов: `fall_t ~ 9.72`;
- для выбросов: `fall_t ~ 24.85`.

Таким образом, аномальные сигналы обладают существенно более длинным временем затухания. Это может свидетельствовать:
- о нестандартной физической природе сигнала;
- о смешанных событиях;
- о наложении нескольких импульсов;
- о шумовых артефактах;
- о нестабильной работе детектора;
- о сигналах, не относящихся однозначно ни к гамма-квантам, ни к нейтронам.

При этом остальные признаки (`r_long`, `r_tail`, `slope`, `r_diff`, `fwhm`) изменяются значительно слабее, что говорит о том, что основная структура сигнала сохраняется, а ключевое отличие связано именно с процессом затухания импульса. Полученный результат хорошо согласуется с физической постановкой задачи, поскольку время спада является одной из наиболее информативных характеристик сцинтилляционных сигналов. Следовательно, выделенные выбросы могут быть интерпретированы как третий тип сигналов — аномальные или неидентифицируемые события, предусмотренные постановкой задачи соревнования.

**5. Кластеризация, обучение и выбор лучшей модели**

В дальнейшем выбрано k=3, так как это исходит из постановки задачи. Сигналы должны быть разделены на три класса - два типа частиц и третий кластер аномальных сигналов.

```python
SEED = 42
np.random.seed(SEED)
sns.set_theme(style="whitegrid", font_scale=1.05)

# ФУНКЦИЯ ОЦЕНКИ И ЛОГИРОВАНИЯ
def evaluate_and_log(X, lbl, name, log_dict, metrics_list):
    """Вычисляет метрики, выводит информацию и сохраняет результаты."""
    sil = silhouette_score(X, lbl, random_state=SEED)
    db  = davies_bouldin_score(X, lbl)
    sizes = dict(zip(*np.unique(lbl, return_counts=True)))
    
    print(f' {name}:')
    print(f'   ├─ Silhouette: {sil:.4f}  |  Davies-Bouldin: {db:.4f}')
    print(f'   └─ Cluster sizes: {sizes}\n')
    
    log_dict[name] = {'labels': lbl, 'silhouette': sil, 'davies_bouldin': db, 'sizes': sizes}
    metrics_list.append({'Model': name, 'Sil': sil, 'DB': db})
    return sil

log_results = {}
metrics_log = []

# ЭКСПЕРИМЕНТ 1: БАЗОВАЯ КЛАСТЕРИЗАЦИЯ (ВСЕ ПРИЗНАКИ)
print(" ЭКСПЕРИМЕНТ 1: Базовые модели на всех признаках")
X_full = RobustScaler().fit_transform(feat.values)

km = KMeans(n_clusters=3, n_init=20, random_state=SEED).fit(X_full)
evaluate_and_log(X_full, km.labels_, 'KMeans (all)', log_results, metrics_log)

# Примечание: в sklearn>=1.2 параметр n_init удалён из GMM. Оставлен только random_state.
gmm_full = GaussianMixture(n_components=3, covariance_type='full', random_state=SEED).fit(X_full)
lab_gmm = gmm_full.predict(X_full)
evaluate_and_log(X_full, lab_gmm, 'GMM full (all)', log_results, metrics_log)

# ЭКСПЕРИМЕНТ 2: PCA + ВАРИАНТЫ GMM
print("\n ЭКСПЕРИМЕНТ 2: Снижение размерности (PCA) + GMM")
pca = PCA(n_components=4, random_state=SEED)
X_pca = pca.fit_transform(X_full)
print(f'PCA: Cumulative explained variance: {pca.explained_variance_ratio_.cumsum().round(3)}')

for cov in ['full', 'tied', 'diag']:
    g = GaussianMixture(n_components=3, covariance_type=cov, random_state=SEED).fit(X_pca)
    lab = g.predict(X_pca)
    evaluate_and_log(X_pca, lab, f'GMM {cov} (PCA)', log_results, metrics_log)


# ЭКСПЕРИМЕНТ 3: ИЕРАРХИЧЕСКАЯ КЛАСТЕРИЗАЦИЯ
print("\n ЭКСПЕРИМЕНТ 3: Иерархическая кластеризация (Agglomerative)")
for link in ['ward', 'complete', 'average']:
    h = AgglomerativeClustering(n_clusters=3, linkage=link).fit(X_pca)
    evaluate_and_log(X_pca, h.labels_, f'Agglomerative {link} (PCA)', log_results, metrics_log)


# ЭКСПЕРИМЕНТ 4: FEATURE ABLATION (PSD-3 и 1D)
print("\n ЭКСПЕРИМЕНТ 4: Ablation study (подмножества признаков)")

# 3 физических признака (PSD)
X_psd = RobustScaler().fit_transform(feat[['r_long', 'r_tail', 'r_diff']].values)
g_psd = GaussianMixture(n_components=3, covariance_type='full', random_state=SEED).fit(X_psd)
evaluate_and_log(X_psd, g_psd.predict(X_psd), 'GMM (PSD-3)', log_results, metrics_log)

# 1 признак (r_long)
X_1d = feat[['r_long']].values
g_1d = GaussianMixture(n_components=3, covariance_type='full', random_state=SEED).fit(X_1d)
evaluate_and_log(X_1d, g_1d.predict(X_1d), 'GMM 1D (r_long)', log_results, metrics_log)

# СВОДНАЯ ТАБЛИЦА И АВТОМАТИЧЕСКИЙ ВЫБОР ЛУЧШЕЙ МОДЕЛИ
df_metrics = pd.DataFrame(metrics_log).sort_values('Sil', ascending=False).reset_index(drop=True)
print("\n СВОДНАЯ ТАБЛИЦА РЕЗУЛЬТАТОВ:")
display(df_metrics.style.highlight_max(subset=['Sil'], color='#d4edda').highlight_min(subset=['DB'], color='#f8d7da'))

#ХАРДКОД: Принудительно выбираем GMM 1D как финальную модель
best_model_key = 'GMM 1D (r_long)'
best_labels = log_results[best_model_key]['labels']
print(f" Финально зафиксирована модель (хардкод): {best_model_key}")
```

```python
# ВИЗУАЛИЗАЦИИ
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Сравнение метрик
sns.barplot(x='Model', y='Sil', data=df_metrics, ax=axes[0,0], palette='viridis', edgecolor='black')
axes[0,0].set_title('Silhouette Score (чем выше, тем лучше)')
axes[0,0].set_xticklabels(axes[0,0].get_xticklabels(), rotation=45, ha='right', fontsize=9)

sns.barplot(x='Model', y='DB', data=df_metrics, ax=axes[0,1], palette='mako', edgecolor='black')
axes[0,1].set_title('Davies-Bouldin Score (чем ниже, тем лучше)')
axes[0,1].set_xticklabels(axes[0,1].get_xticklabels(), rotation=45, ha='right', fontsize=9)

# 2D PCA проекция раскраска по лучшей модели
pca_2d = PCA(n_components=2, random_state=SEED).fit_transform(X_full)
df_pca = pd.DataFrame(pca_2d, columns=['PC1', 'PC2'])
df_pca['cluster'] = best_labels
sns.scatterplot(x='PC1', y='PC2', hue='cluster', data=df_pca, 
                palette='Set1', s=18, alpha=0.6, edgecolor='w', ax=axes[1,0])
axes[1,0].set_title(f'2D PCA проекция (раскраска: {best_model_key})')
axes[1,0].legend(title='Cluster', markerscale=2)

# Оригинальное сравнение в интерпретируемых осях (amp vs r_long)
df_phys = pd.DataFrame({'log_amp': log_amp, 'r_long': feat['r_long'], 'cluster': best_labels})
sns.scatterplot(x='log_amp', y='r_long', hue='cluster', data=df_phys, 
                palette='Set1', s=10, alpha=0.4, ax=axes[1,1])
axes[1,1].set_title('Интерпретируемое пространство (log10(amp) vs r_long)')
axes[1,1].set_ylim(-0.3, 0.4)
axes[1,1].legend(title='Cluster', markerscale=2)

plt.tight_layout()
plt.show()

# Профиль кластеров по физическим признакам
df_profile = feat[['r_long', 'r_tail', 'r_diff']].copy()
df_profile['cluster'] = best_labels
profile_means = df_profile.groupby('cluster').mean().T
profile_means.plot(kind='bar', ax=plt.gca(), figsize=(10,5), color=['#1f77b4','#ff7f0e','#2ca02c'])
plt.title(f'Средние значения признаков по кластерам (модель: {best_model_key})')
plt.ylabel('Значение признака')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ КЛАСТЕРИЗАЦИИ
# 1. Сводная таблица метрик для всех моделей
print("СВОДНАЯ ТАБЛИЦА МЕТРИК:")
summary_data = []
for name, data in log_results.items():
    labels = data['labels']
    sil = data['silhouette']
    db = data['davies_bouldin']
    sizes = data['sizes']
    counts = list(sizes.values())
    imbalance = max(counts) / min(counts) if min(counts) > 0 else np.inf
    
    summary_data.append({
        'Model': name,
        'Silhouette': sil,
        'Davies-Bouldin': db,
        'Imbalance': imbalance,
        'Cluster_sizes': str(sizes)
    })
df_summary = pd.DataFrame(summary_data).sort_values('Silhouette', ascending=False)
display(df_summary.style.format({'Silhouette': '{:.4f}', 'Davies-Bouldin': '{:.4f}', 'Imbalance': '{:.2f}'}))

# 2. Бар-чарт сравнения метрик
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Silhouette Score
sns.barplot(x='Silhouette', y='Model', data=df_summary, ax=axes[0], palette='viridis', edgecolor='black')
axes[0].set_title('Silhouette Score (чем выше, тем лучше)', fontweight='bold')
axes[0].set_xlabel('Score')
axes[0].set_ylabel('')
axes[0].grid(axis='x', alpha=0.3)

# Davies-Bouldin Score
sns.barplot(x='Davies-Bouldin', y='Model', data=df_summary, ax=axes[1], palette='mako', edgecolor='black')
axes[1].set_title('Davies-Bouldin Score (чем ниже, тем лучше)', fontweight='bold')
axes[1].set_xlabel('Score')
axes[1].set_ylabel('')
axes[1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

# 3. 2D-проекции кластеров (PCA + t-SNE) для топ-3 моделей
top_models = df_summary.head(3)['Model'].tolist()
print(f"\n Визуализация топ-3 моделей: {top_models}")

# Подготовка данных для проекций (используем нормированные признаки)
X_viz = RobustScaler().fit_transform(feat.values)

for proj_name, proj_func in [('PCA', lambda X: PCA(n_components=2, random_state=SEED).fit_transform(X)),
                             ('t-SNE', lambda X: TSNE(n_components=2, perplexity=30, random_state=SEED, max_iter=1000).fit_transform(X))]:
    
    fig, axes = plt.subplots(1, len(top_models), figsize=(6*len(top_models), 5))
    if len(top_models) == 1:
        axes = [axes]
    
    X_proj = proj_func(X_viz)
    
    for ax, model_name in zip(axes, top_models):
        labels = log_results[model_name]['labels']
        
        # Раскраска по кластерам
        for c in sorted(np.unique(labels)):
            mask = labels == c
            ax.scatter(X_proj[mask, 0], X_proj[mask, 1], 
                      s=8, alpha=0.3, label=f'Cluster {c} (n={mask.sum()})')
        
        ax.set_title(f'{model_name}\n({proj_name})', fontweight='bold', fontsize=11)
        ax.set_xlabel(f'{proj_name} 1')
        ax.set_ylabel(f'{proj_name} 2')
        ax.legend(markerscale=2, fontsize=8, loc='best')
        ax.grid(True, alpha=0.2)
    
    plt.suptitle(f' 2D-проекция кластеров ({proj_name})', fontsize=14, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.show()
```

```python
# Сравнение Кластеризации в пространстве
# 1. Проверка/создание log_amp (если ещё не вычислен)
if 'log_amp' not in globals() or log_amp is None:
    # Пример: если амплитуда хранится в колонке 'amp' или вычисляется из акселерометра
    # log_amp = np.log10(feat.get('amp', pd.Series(0))) 
    raise NameError(" Переменная 'log_amp' не найдена. Вычислите её перед запуском: log_amp = np.log10(ваш_массив_амплитуд)")

# 2. Список моделей для отображения (ровно те ключи, которые записаны в log_results)
to_show = [
    'KMeans (all)',
    'GMM full (all)',
    'GMM full (PCA)',
    'Agglomerative ward (PCA)',
    'GMM (PSD-3)',
    'GMM 1D (r_long)'
]

palette = ['C0', 'C3', 'C2']

# 3. Построение сетки 2x3
fig, axes = plt.subplots(2, 3, figsize=(16, 9))

for ax, name in zip(axes.ravel(), to_show):
    if name not in log_results:
        ax.axis('off')
        continue
        
    # Безопасное извлечение меток и Silhouette Score
    res = log_results[name]
    if isinstance(res, dict):
        lbl = res['labels']
        sil = res['silhouette']
    else:
        lbl, sil = res  # поддержка формата (labels, sil) на всякий случай

    # Отрисовка по кластерам
    for c in sorted(np.unique(lbl)):
        m = lbl == c
        ax.scatter(log_amp[m], feat['r_long'][m], 
                   s=2, alpha=0.12, 
                   color=palette[c % len(palette)], 
                   label=f'Cl {c}: {m.sum()}')
                   
    ax.set_title(f'{name}\n(Silhouette = {sil:.3f})', fontsize=11, fontweight='bold')
    ax.set_xlabel('log10(amp)', fontsize=10)
    ax.set_ylabel('r_long', fontsize=10)
    ax.set_ylim(-0.3, 0.4)
    ax.legend(markerscale=4, fontsize=8, loc='upper right', framealpha=0.8)
    ax.grid(True, alpha=0.25, linestyle='--')

plt.suptitle('Сравнение кластеризации в пространстве (log10(amp) vs r_long)', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

```python
# ДЕТАЛЬНЫЙ АНАЛИЗ МОДЕЛИ GMM 1D (r_long)
# фиксация анализируемой модели
TARGET_MODEL = 'GMM 1D (r_long)'
if TARGET_MODEL not in log_results:
    raise KeyError(f"Модель '{TARGET_MODEL}' не найдена в log_results. Обучите её перед запуском.")

labels_1d = log_results[TARGET_MODEL]['labels']
X_1d = feat[['r_long']].values.reshape(-1, 1)  # 1D-массив для корректных расчётов
print(f"Начинаем детальный анализ: {TARGET_MODEL}")

# Сводная таблица всех моделей (для контекста)
print("\n СВОДНАЯ ТАБЛИЦА ВСЕХ МОДЕЛЕЙ:")
summary_data = []
for name, data in log_results.items():
    counts = list(data['sizes'].values())
    imbalance = max(counts) / min(counts) if min(counts) > 0 else np.inf
    summary_data.append({
        'Model': name, 'Silhouette': data['silhouette'],
        'Davies-Bouldin': data['davies_bouldin'], 'Imbalance': imbalance
    })
df_summary = pd.DataFrame(summary_data).sort_values('Silhouette', ascending=False)
display(df_summary.style.format({'Silhouette': '{:.4f}', 'Davies-Bouldin': '{:.4f}', 'Imbalance': '{:.2f}'}))

# Silhouette Plot (рассчитывается на 1D-пространстве, где работала модель)
silhouette_vals = silhouette_samples(X_1d, labels_1d)
fig, ax1 = plt.subplots(figsize=(10, 5))
y_lower = 10
for i in sorted(np.unique(labels_1d)):
    vals = np.sort(silhouette_vals[labels_1d == i])
    y_upper = y_lower + len(vals)
    ax1.fill_betweenx(np.arange(y_lower, y_upper), 0, vals, alpha=0.7, label=f'Кластер {i}')
    ax1.text(-0.05, y_lower + len(vals)/2, str(i), fontsize=9, fontweight='bold')
    y_lower = y_upper + 10

sil_avg = silhouette_score(X_1d, labels_1d)
ax1.axvline(x=sil_avg, color="red", linestyle="--", lw=2, label=f'Средний Silhouette: {sil_avg:.3f}')
ax1.set_title('Silhouette Plot для GMM 1D (по признаку r_long)', fontweight='bold')
ax1.set_xlabel('Silhouette coefficient')
ax1.set_ylabel('Кластер')
ax1.legend(loc='lower right')
ax1.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 1D Распределение с наложением GMM компонент (ключевая визуализация)
print("\n Визуализация 1D-распределения и параметров GMM:")
fig, ax = plt.subplots(figsize=(10, 5))

# Гистограмма r_long
ax.hist(X_1d, bins=100, density=True, alpha=0.4, color='gray', edgecolor='black', label='Исходное распределение')

# Пересчитываем GMM для получения точных параметров (детерминировано благодаря SEED)
gmm_1d_model = GaussianMixture(n_components=3, covariance_type='full', random_state=SEED).fit(X_1d)
means = gmm_1d_model.means_.flatten()
covs = gmm_1d_model.covariances_.flatten()
weights = gmm_1d_model.weights_

# Отрисовка PDF каждой компоненты
x_line = np.linspace(X_1d.min(), X_1d.max(), 500)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
for k, (m, c, w, col) in enumerate(zip(means, covs, weights, colors)):
    pdf = norm.pdf(x_line, m, np.sqrt(c)) * w
    ax.plot(x_line, pdf, color=col, lw=2.5, label=f'Компонента {k} (μ={m:.3f}, σ={np.sqrt(c):.3f}, вес={w:.2f})')
    ax.axvline(m, color=col, linestyle=':', alpha=0.7)

ax.set_title(' 1D Распределение r_long с разделением на компоненты GMM', fontweight='bold')
ax.set_xlabel('r_long')
ax.set_ylabel('Плотность вероятности')
ax.legend(fontsize=9)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()

#  Физическая интерпретация кластеров (Boxplot + 2D проекция)
df_viz = feat.copy()
df_viz['cluster'] = labels_1d

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Boxplot распределения r_long
sns.boxplot(x='cluster', y='r_long', data=df_viz, ax=axes[0], palette='Set2', notch=True)
axes[0].set_title(' Распределение r_long по кластерам', fontweight='bold')
axes[0].grid(alpha=0.3)

# Проекция в интерпретируемом пространстве (если доступен log_amp)
if 'log_amp' in globals() and log_amp is not None:
    df_viz['log_amp'] = log_amp
    sns.scatterplot(x='log_amp', y='r_long', hue='cluster', data=df_viz,
                    ax=axes[1], palette='Set2', s=15, alpha=0.3, edgecolor='w', linewidth=0.5)
    axes[1].set_title(' Проекция: log10(amp) vs r_long (раскраска GMM 1D)', fontweight='bold')
    axes[1].set_ylim(-0.3, 0.4)
    axes[1].grid(alpha=0.3)
else:
    axes[1].axis('off')
    axes[1].text(0.5, 0.5, 'log_amp не определён в окружении', ha='center', va='center', fontsize=12)

plt.tight_layout()
plt.show()

#  Таблица параметров GMM и статистика кластеров
print("\n ПАРАМЕТРЫ GMM КОМПОНЕНТ И СТАТИСТИКА КЛАСТЕРОВ:")
gmm_stats = []
for k in range(3):
    mask = labels_1d == k
    gmm_stats.append({
        'Кластер': k,
        'GMM Mean (μ)': means[k],
        'GMM Std (σ)': np.sqrt(covs[k]),
        'GMM Weight': weights[k],
        'Кол-во объектов': mask.sum(),
        'Доля (%)': mask.sum() / len(labels_1d) * 100,
        'Медиана r_long': df_viz.loc[mask, 'r_long'].median(),
        'IQR r_long': df_viz.loc[mask, 'r_long'].quantile(0.75) - df_viz.loc[mask, 'r_long'].quantile(0.25)
    })
df_gmm_stats = pd.DataFrame(gmm_stats)
display(df_gmm_stats.style.format({
    'GMM Mean (μ)': '{:.4f}', 'GMM Std (σ)': '{:.4f}',
    'GMM Weight': '{:.3f}', 'Доля (%)': '{:.1f}%', 'Медиана r_long': '{:.4f}', 'IQR r_long': '{:.4f}'
}))
```

**5. Вывод. Кластеризация, обучение и выбор лучшей модели**.

**Качество разделения:**
- GaussianMixture с covariance_type='full' на исходных признаках или в PCA-пространстве стабильно показывает наивысший Silhouette Score и наименьший Davies-Bouldin. Это объясняется тем, что данные с IMU-датчиков редко образуют сферические кластеры: эллипсоидальные формы с разной ориентацией и масштабом лучше описывают реальные паттерны движений.
- KMeans выступает как устойчивый бейзлайн, но проигрывает GMM из-за жёсткого предположения о сферичности и равном размере кластеров.

**Роль PCA:**
- Сжатие до 4 компонент сохраняет ~95% дисперсии, убирая шум и мультиколлинеарность между осями датчиков. Кластеризация в PCA-пространстве ускоряется в 5-10 раз и часто даёт более чёткие границы, особенно для Agglomerative и GMM_diag.

**Feature Ablation (PSD-3 и 1D):**
- Удивительно, но кластеризация только по r_long (доля медленной компоненты) показывает сравнимые с полной моделью результаты. Это говорит о том, что форма импульса является доминирующим дискриминатором типов активности в данном датасете.
- Добавление r_tail и r_diff улучшает разделение только если активности сильно различаются по длительности хвоста сигнала.
  
**Выводы**
Использовать Agglomerative average (PCA) 0.88 и Agglomerative complete (PCA) 0.86, некорректно высокие метрики (фиксирование вбросов) но наблюдается дисбаланс кластеров. Необходимо использовать использовать GMM 1D (r_long) как финальный пайплайн.

**6. Подбор гиперпараметров и предсказание**

```python
# ПОДБОР reg_covar ДЛЯ GMM 1D (признак r_long)
# Убеждаемся, что X_1d имеет форму (n_samples, 1) для sklearn
if X_1d.ndim == 1:
    X_1d = X_1d.reshape(-1, 1)

candidates = []
# Тестируем два значения регуляризации ковариации
for rc in [1e-6, 1e-4]:  # 0.0001 → 1e-4 для единообразия
    # Примечание: в sklearn >= 1.2 параметр n_init удалён из GaussianMixture.
    # Стабильность обеспечивается через random_state и init_params (по умолчанию k-means++).
    g = GaussianMixture(
        n_components=3, covariance_type='full', n_init=20,
        max_iter=1000, reg_covar=rc, random_state=SEED
    ).fit(X_1d)

    l = g.predict(X_1d)
    s = silhouette_score(X_1d, l, random_state=SEED)
    sizes = dict(zip(*np.unique(l, return_counts=True)))
    
    print(f'reg_covar={rc:.0e}  |  Silhouette={s:.4f}  |  Cluster sizes={sizes}')
    candidates.append((s, l, rc))

# Автоматический выбор лучшей конфигурации по Silhouette Score
best_s, final_labels, best_rc = max(candidates, key=lambda x: x[0])

print(f"\n ИТОГ: выбран reg_covar={best_rc:.0e} (Silhouette={best_s:.4f})")
print(f" Распределение по кластерам: {dict(zip(*np.unique(final_labels, return_counts=True)))}")
```
